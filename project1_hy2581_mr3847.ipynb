### Data Visualization

#### a) Do the plantform with higher average sales on games have higher average users score?

#### The sales information of different video game platform .

```python
# group the dataset by different platforms and calculate the mean of global sales for each platform, 
# then rank the result in descending order.
df_platform_mean = df.groupby('Platform').mean().sort_values('Global_Sales',ascending = False).drop(['Critic_Score','Critic_Count','User_Count'],axis = 1)['Global_Sales']
```

```python
# plot a bar chart for the average global sales for different platform in descending order.
plt.figure(figsize=(15, 5))
df_platform_mean.plot(kind = 'bar')
plt.ylabel('Average Sales')
plt.xlabel('Platforms')
```

#### The score of video games from different platform

```python
# create a new dataset that contained 'Platform','Global_Sales' and 'User_Score', then drop the rows whose user score is recorded as 'tbd'
test = df[['Platform','Global_Sales','User_Score']].dropna(how = 'any',axis=0)
test1 = test[test['User_Score'] != 'tbd']
```

```python
# change the type of data in column 'User_Score' from string to float.
test1.loc[:,['User_Score']] = pd.to_numeric(test1['User_Score'])
test1
```

```python
# calculate the mean user score for different platform and drop the rows that the user score is missing
df_userscore = test1.groupby('Platform')['User_Score'].mean()
df_userscore = df_userscore.dropna(how = 'any',axis = 0).sort_values(ascending = False)
df_userscore
```

```python
# plot a bar chart for the average user score for different platform in descending order.
plt.figure(figsize=(15,5))
df_userscore.plot(kind = 'bar')
plt.ylabel('Average User Score')
plt.xlabel('Platforms')
```

#### The relationship between global sales and user score.

```python
# calculate the correlation coefficient for global sales and user score.
test1.corr()
```

#### b) The percentage of sales of each genre over years

```python
# calculate the proportion of different game genres for global sales of each year.
global_sale_year = (df.groupby(['Year_of_Release','Genre']).Global_Sales.sum()/df.groupby('Year_of_Release').Global_Sales.sum())*100
```

```python
# make a plot of the proportion of different game genre alone years.  è¿™ç§å›¾æˆ‘å¿˜è®°å«å•¥äº†

global_sale_year.unstack().plot(kind='area',stacked=True,colormap = 'Spectral',figsize = (13,4))
plt.legend(bbox_to_anchor=(1.05, 0), loc=3, borderaxespad=0)
plt.ylabel('Percentage(%)')
plt.xlabel('Years')
```

#### c) Find the best sales platform based on the 7th generation and 8th generation

     7th generation

```python
# create a new dataset that contained the rows that their platform is Wii or PS3 or X360.
video7th = df[(df['Platform'] == 'Wii') | (df['Platform'] == 'PS3') | (df['Platform'] == 'X360')]
video7th
```

```python
# calculate the total global sales of 3 different platforms of each year and make a stacked bar plot
yearlySales = video7th.groupby(['Year_of_Release','Platform']).Global_Sales.sum()
yearlySales = yearlySales.unstack()
yearlySales.plot(kind='bar',width = 0.8,figsize=(15, 5))
plt.ylabel('Global Sales')
plt.xlabel('Years')
```

    8th generation

```python
video8th = df[(df['Platform'] == 'WiiU') | (df['Platform'] == 'PS4') | (df['Platform'] == 'XOne')]
```

```python
yearlySales = video8th.groupby(['Year_of_Release','Platform']).Global_Sales.sum()
yearlySales = yearlySales.unstack()
yearlySales.plot(kind='bar',width = 0.8,figsize=(15, 5))
plt.ylabel('Global Sales')
plt.xlabel('Years')
```

```python
# pie chart of global sales 
plt.style.use('seaborn-white')
colors = ['lavender','skyblue','lightsteelblue']
plt.figure(figsize=(20,9.5))
plt.subplot(121)
plt.pie(
   video7th.groupby('Platform').Global_Sales.sum(),
    labels=video7th.groupby('Platform').Global_Sales.sum().index,
    shadow=False,
    colors=colors,
    explode=(0.05, 0.05, 0.05),
    startangle=90,
    autopct='%1.1f%%')
plt.title('Pie Chart of Global Sales for 7th generation')
plt.subplot(122)
plt.pie(
   video8th.groupby('Platform').Global_Sales.sum(),
    labels=video8th.groupby('Platform').Global_Sales.sum().index,
    shadow=False,
    colors=colors,
    explode=(0.05, 0.05, 0.05),
    startangle=90,
    autopct='%1.1f%%')
plt.title('Pie Chart of Global Sales for 8th generation')
plt.show()
```
